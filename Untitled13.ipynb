```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

```

# Fuel Efficiency Prediction

This data set has 398 rows and 9 columns and provides mileage, horsepower, model year, and other technical specifications for cars.

- mpg : fuel efficiency measured in miles per gallon (mpg)
- cylinders: number of cylinders in the engine
- displacement: engine displacement (in cubic inches)
- horsepower: engine horsepower
- weight: vehicle weight (in pounds)
- acceleration: time to accelerate from O to 60 mph (in seconds)
- model year: model year
- origin: origin of car (1: American, 2: European, 3: Japanese)
- car name: car name

```python
df = pd.read_csv('/content/auto-mpg.csv')
df.head()
```

```python
df.info()
```

```python
#convert the '?' values to NaN
df['horsepower'].replace('?',np.nan,inplace=True)
```

```python
# convert into int
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
#calculate the mean of the column and replace NaN values with the mean
mean_horsepower = df['horsepower'].mean()
df['horsepower'].fillna(mean_horsepower, inplace=True)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.describe().T
```

drop the column 'car name'

```python
df.drop(columns=['car name'],inplace=True)
```

```python
df.describe().T
```

#split

```python
X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
 from sklearn.model_selection import train_test_split
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# Feature engineering & Model Creation

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
# Create a ColumnTransformer to handle missing values and scaling
transformers = [
    ('num', StandardScaler(), X.columns)
]
```

```python
preprocessor = ColumnTransformer(transformers)
```

```python
# Create a pipeline that combines data preprocessing and linear regression
model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
# Fit the model on the training data
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
prediction_df = pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference / Residual':y_test-y_pred})
prediction_df
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# The coefficient of determination: 1 is perfect predictio
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
print("Mean absolute error: %.2f" % mean_absolute_error(y_test,y_pred))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
plt.scatter(y_test,y_pred);
plt.xlabel('Actual');
plt.ylabel('Predicted');
```

```python
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```
